```python
import pandas as pd
import numpy as np
import os
```

```python
folder_name = 'orderline_price_asin_small'
```

```python
### load_xml_query
xml_query_file_path = "data/" + folder_name + "/" + "XML_query.dat"
with open(xml_query_file_path) as f:
    content = f.readlines()
content = [x.strip() for x in content]

all_elements_name = content[0].split(" ")

relationship_matrix = np.zeros((len(all_elements_name), len(all_elements_name)))

for i in range(1, len(content)):
    relationship = content[i].split(" ")
    relationship_matrix[all_elements_name.index(relationship[0]), all_elements_name.index(relationship[1])] = int(relationship[2])
```

```python
### load_tables
tables_name = []
tables_content = {}
for file_name in os.listdir("./data/" + folder_name):
    if 'table' in file_name:
        table_name = file_name[:-10]
        tables_name.append(table_name)
        tables_content[table_name] = pd.read_csv('./data/' + folder_name + '/' + file_name, names = table_name.split('_'), sep='\s+')
```

```python
### load elements
elements_value = {}
for element_name in all_elements_name:
    path = "data/" + folder_name + "/" + element_name + "_v.dat"
    elements_value[element_name] = pd.read_csv(path, names = [element_name], sep='\s+')
elements_value
```

```python
def print_statistical_overview(series, tab_index = 1):
    print('\t' * tab_index, 'n:', len(series.as_matrix()))
    print('\t' * tab_index, 'mean:', np.mean(series.as_matrix()))
    print('\t' * tab_index, 'std:', np.std(series.as_matrix()))
```

```python
print('ELEMENTS: ', ', '.join(all_elements_name))
print('TABLES: ', ', '.join(tables_name))
```

```python
for element in all_elements_name:
    print('XML overview:', element)
    print_statistical_overview(elements_value[element][element])
    print('Table containing element')
    for table in tables_name:
        if element in table:
            print('\t', table)
            print_statistical_overview(tables_content[table][element], 2)
```
